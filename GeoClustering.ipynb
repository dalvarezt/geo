### Procesamiento y análisis de los datos:

```python
Dist_dia = result_df.groupby(["FECHA"]).size().reset_index()
Dist_dia.columns = ["FECHA", "TotalRegistros"]
Dist_dia
```

```python
Filtro_Fecha = '2018-08-01'
Dist_dia_Modelo = result_df.loc[result_df['FECHA']== '2018-08-01']
```

## MODELAMIENTO 
### Generación del modelo de clúster geográfico

```python
points = np.array(Dist_dia_Modelo[['LATITUD_INICIO', 'LONGITUD_INICIO']])
rads_points = np.radians(points)
clusterer = hdbscan.HDBSCAN(min_cluster_size= 20, min_samples = 2, metric='haversine', prediction_data=True)
cluster_labels = clusterer.fit_predict(rads_points)
Dist_dia_Modelo["clusters"] = cluster_labels 
```

```python
Dist_dia_Modelo
```

```python
resumenCluster = Dist_dia_Modelo.groupby("clusters").agg({'TRIPID' : ['count'], 'LATITUD_INICIO':['mean'], 'LONGITUD_INICIO' : ['mean']})
resumenCluster= pd.DataFrame(resumenCluster.to_records())
resumenCluster.columns = ["cluster", "total","Latitude_Cent","Longitude_Cent"]
resumenCluster["Fecha"] = Filtro_Fecha
resumenCluster.sort_values(by = "total", ascending=False)
resumenCluster = resumenCluster.loc[resumenCluster["cluster"] >= 0]
resumenCluster
```

### Evaluación de los Resultados

```python
#creating a map that's centered to our sample
df_ClusterSum1 = resumenCluster

map_Alert36 = folium.Map(location=[df_ClusterSum1['Latitude_Cent'].mean(), 
                                df_ClusterSum1['Longitude_Cent'].mean()], 
                                zoom_start=7)

for row in df_ClusterSum1.itertuples():
    map_Alert36.add_child(folium.Circle(location=[row.Latitude_Cent,row.Longitude_Cent],
                                        popup=row.total, 
                                        radius = row.total
                                       ))

map_Alert36
```
